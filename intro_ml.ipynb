```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Definer parametre for hver normalfordeling
# TODO: Ændre disse for at få forskellige fordelinger...
blaa_fordeling = [0, 1, "blue"]
groen_fordeling = [5, 2, "green"]
red_fordeling = [-2, 0.5, "red"]

# Lav x værdier, bare de værdier vi plotter langs...
x_vals = np.linspace(-10, 10, 1000)

plt.figure(figsize=(10, 6))

for mu, sigma, color  in [blaa_fordeling, groen_fordeling, red_fordeling]:
    # Generér sandsynligheder fra fordeling
    probabilities = stats.norm.pdf(x_vals, mu, sigma)
    label = f"μ = {mu}, σ² = {sigma}"
    
    # Plot sandsynligheder
    plt.plot(x_vals, probabilities, color=color, label=label)

# Plotting ting
plt.title("Normalfordelinger")
plt.xlabel("x")
plt.ylabel("Densitet")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
